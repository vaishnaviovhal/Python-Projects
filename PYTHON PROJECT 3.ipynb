TASK2:QUESTION 1-T-Test

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("TWTR.xlsx")
df.head()
```

```python
df.columns
```

```python
df = df[['High', 'Close']].dropna()
```

```python
from scipy.stats import ttest_ind

t_stat, p_val = ttest_ind(df['High'], df['Close'])

print("T-statistic:", t_stat)
print("P-value:", p_val)
```

TASK3:QUESTION 1-Line Plot of Closing Price

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Close'])
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("Twitter Closing Price Over Time")
plt.show()
```

QUESTION 2-Volume Traded Over Time

```python
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Volume'])
plt.xlabel("Date")
plt.ylabel("Volume")
plt.title("Trading Volume Over Time")
plt.show()
```

QUESTION 3-High vs Low Price Comparison

```python
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['High'], label='High')
plt.plot(df['Date'], df['Low'], label='Low')
plt.legend()
plt.title("High vs Low Prices")
plt.show()
```

QUESTION 4-Histogram of Closing Prices

```python
plt.hist(df['Close'], bins=20)
plt.xlabel("Closing Price")
plt.ylabel("Frequency")
plt.title("Distribution of Closing Prices")
plt.show()
```

QUESTION 5-Moving Average

```python
df['MA20'] = df['Close'].rolling(window=20).mean()

plt.figure(figsize=(10,5))
plt.plot(df['Close'], label='Close Price')
plt.plot(df['MA20'], label='20-Day MA')
plt.legend()
plt.title("20-Day Moving Average")
plt.show()
```

QUESTION 6-Box Plot

```python
plt.boxplot(df[['Open', 'High', 'Low', 'Close']])
plt.xticks([1,2,3,4], ['Open','High','Low','Close'])
plt.title("Stock Price Distribution")
plt.show()
```
