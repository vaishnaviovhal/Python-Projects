```python
import pandas as pd

df = pd.read_excel("WorldUniversity.xlsx")
df.head()
```

```python
df.shape        # rows, columns
```

```python
df.info()       # column names + datatypes
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df['country'] = df['country'].str.strip()
```

```python
num_cols = df.columns.drop(['institution', 'country'])
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors='coerce')
```

```python
len(df)
```

```python
df.shape[1]
```

```python
df[(df.country=='United Kingdom') & (df.year==2013) & (df.score>80)]
```

```python
df.groupby('country').size()
```

```python
df[df.year==2014].groupby('country')['score'].mean()
```

```python
top10 = df[df.year==2012].sort_values('score', ascending=False).head(10)

plt.figure()
plt.bar(top10['institution'], top10['score'])
plt.xticks(rotation=90)
plt.title("Top 10 Universities by Score - 2012")
plt.show()
```

```python
plt.figure()
plt.scatter(df['quality_of_faculty'], df['quality_of_education'])
plt.xlabel("Quality of Faculty")
plt.ylabel("Quality of Education")
plt.show()
```

```python
df.to_csv("Cleaned_World_University_Data.csv", index=False)
```
